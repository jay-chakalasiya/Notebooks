### Generic Equation

##### Y(W1*X1 + W2*X2 + b) + Q >=1

##### for first iteration q=0, and we can select 3 equation for get the solution of W1, W2 and b

## Functional Variables
### arr = points, 
### bias = bias array, 
### classs = y value, 
### ans = RHS of equation = 1
### seed = seed value to randomaly select 3 equation

```python
import numpy as np
arr = np.array([[0,0],[0,1],[1,0],[1,1],[-1,-1]])
bias = np.array([[1,1,1,1,1]])
classs = np.array([-1,1,1,-1,-1])
ans = np.array([1,1,1,1,1])
```

```python
def solveeq(arr, bias, classs, ans, seed):
    
    # forming equation of form w1*px + w2*py +b
    arr = np.concatenate((arr, bias.T), axis=1)
    
    # y(w1*px + w2*py +b)
    for i in range(len(arr)):
        arr[i] = arr[i]*classs[i]
    
    # solving y(w1*px + w2*py +b) = 1 for any three equations to get w1, w2, b
    np.random.seed(seed)
    eqs = []
    while len(eqs)<3:
        i = np.random.randint(0,len(arr))
        if i not in eqs:
            eqs.append(i)
            
    w1,w2,b = np.linalg.solve(np.array([arr[eqs[0]],arr[eqs[1]],arr[eqs[2]]]), np.array([ans[eqs[0]],ans[eqs[1]],ans[eqs[2]]]))
    
    
    # getting residuals = q errors
    qvals = arr @ np.array([w1,w2,b])
    
    # finding missing weights
    index=[]
    Ws = []
    for i in range(len(qvals)):
        if qvals[i]!=1:
            index.append(i)
            Ws.append((-1*(qvals[i]-ans[i]))/classs[i])
        else:
            Ws.append(0)
    
    return arr, eqs, w1, w2, b, qvals, index, Ws
```

```python
def printans(arr, bias, classs, ans):
    
    for seed in range(100):
        try:
            arr, eqs, w1, w2, b, qvals, index, Ws = solveeq(arr, bias, classs, ans, seed)
            print("-------------------- seed value = ",seed," --------------------")
            print("")
            print("first solved for equations no --  ",eqs)
            for i in range(len(eqs)):
                print(classs[i],"*( w1*",arr[eqs[i]][0]," + w2*",arr[eqs[i]][1]," + b) = ", ans[i])
            print("")
            print("== solution == ")
            print("W1 = ", w1, "      w2 = ",w2,"      b = ",b)
            print("")
            print("Complete solution after finding extra weights")

            for i in range(len(arr)):
                if i not in index:
                    print(classs[i],"*( ",w1,"*",arr[i][0]," + ", w2,"*",arr[i][1]," + ", b," ) = ", ans[i])
                else:
                    print(classs[i],"*( ",w1,"*",arr[i][0]," + ", w2,"*",arr[i][1]," + ",Ws[i]," + ", b," ) + ", qvals[i]," = ", ans[i])
        except:
            print("-------------------- seed value = ",seed," --------------------")
            print("can't be done for ",seed)
```

```python
printans(arr, bias, classs, ans)
```
